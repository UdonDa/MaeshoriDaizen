# 2-2 条件指定による抽出
### 条件によるデータ行の抽出

```python
# Not Awesome
# 配列に条件式を指定して、match
# -> 変更に弱い
reserve_tb[(reserve_tb['checkout_date'] >= '2016-10-3') & (reserve_tb['checkout_date'] <= '2016-10-4')].iloc[0:5, :]
```

```python
# Not Awesome
# loc関数のⅠ次元目に条件指定して、match
# -> 長いし変更だるい
reserve_tb.loc[(reserve_tb['checkout_date'] >= '2016-10-3') & (reserve_tb['checkout_date'] <= '2016-10-4'), :].iloc[0:5, :]
```

```python
# Awesome
# query関数
# -> 可読性よい！
reserve_tb.query('"2016-10-3" <= checkout_date <= "2016-10-4"').iloc[0:5, :]
```

# 2-3 データ値に基づかないサンプリング
### ランダムサンプリング

```python
# Awesome
# pandas sample関数
reserve_tb.sample(frac=0.5).iloc[0:5, :]
```

```python
reserve_tb.sample(n=5)
```

# 2-4 集約IDに基づくサンプリング
顧客IDに対してランダムサンプリングしてから、サンプリングした顧客IDの予約レコードのみを抽出

### IDごとのサンプリング

```python
# Awesome
# unique関数でcustomer_id一覧作成後, sample0.5 -> isin関数で一致する行を抽出
# -> query関数はinがない
target = pd.Series(reserve_tb['customer_id'].unique()).sample(frac=0.5)
reserve_tb[reserve_tb['customer_id'].isin(target)].iloc[0:10, :]
```
